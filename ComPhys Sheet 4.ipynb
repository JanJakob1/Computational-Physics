```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_k(x):
    return (epsilon-x)

def num(x0, y0, y1, h, n):
    x=np.array([x0, x0+h])
    y=np.array([y0, y1])
    for i in range(1, n+1):
        k0=get_k(x0+(i-1)*h)
        k1=get_k(x0+i*h)
        k2=get_k(x0+(i+1)*h)
        y2=(2*(1-5/12*h**2*k1)*y1-(1+1/12*h**2*k0)*y0)/(1+1/12*h**2*k2)
        y0=y1
        y1=y2
        x=np.append(x, x0+(i+1)*h)
        y=np.append(y, y1)
    return x,y
```

# 2) Hausaufgabe
Für die neutrinos ist das Potetial gegeben mit $$V(z)=mgz$$ für z>=0. Die Bei z=0 sollen dei Neutrinos gepsiegelt werden indem $V(z)=\infty$ für z<0. Man kann die Schrödingergleichung $$\Psi'' (z) + \frac{2m}{\hbar^2}(E-V(z))\Psi(z)=0$$ mit Hilfe der Subtstitutionen
$$\epsilon :=E/E_0, x=z/z_0$$

$$z_0=\frac{\hbar^2}{2m^2g}$$

$$E_0=\frac{\hbar ^2}{2m}$$
umschreiben zu 
$$\Psi'' (x) + (\epsilon-x)\Psi(x)=0$$

Die Integration wird gestartet mit $\Psi (0)=0$ und $\Psi(h)=h$

## 2.1

```python
# Stepsize
h=0.01

# Anfangswerte
psi_0=0
psi_h=h
# wie weit solle x propagieren
x0=0
x_max=30

# anzahl an benötigten Schritten
m=int(x_max/h)
```

```python
# Plotten für 2 verschiedene Epsilons
epsilon=2.33
x,y=num(x0, psi_0, psi_h, h, m)
plt.plot(x,y)
plt.title("Epsilon = %.4f" %epsilon)
plt.show()
epsilon=2.34
x,y=num(x0, psi_0, psi_h, h, m)
plt.plot(x,y)
plt.title("Epsilon = %.4f" %epsilon)
plt.show()
```

Hier sind die Plots für $\epsilon=2.33$ und $\epsilon=2.34$ dargestellt. Für $\epsilon=2.33$ ist die Asymptote positiv, für $\epsilon=2.34$ negativ.

## 2.2
Für die Eigenwerte $\epsilon_n$ von der Schrödingergleichung gibt es normalisierbare Eigenfunktionen mit $\psi(x) \rightarrow 0$ für $x \rightarrow \infty$. Das bedeutet, dass das Vorzeichen der Funtkion $\psi(x)$ für $x \rightarrow \infty$ das Vorzeichen wechselt, wenn man von kleineren zu größeren Werten um die Eigenwerte variiert. Im folgenden wurden die ersten 3 Eigenwerte auf 2 Stellen genau durch Beobachten der Asymtopten für verschiedene $\epsilon$ bestimmt. 

```python
# hier werden um einen Mittelwert von epsilon im Bereich von Mittelwert-bereich bis Mittelwert+bereich Plots 
# in der Quantität anzahl (+-1 durch mögliche Rundungen) mit gleichmäßigen Abständen der epsilon geplottet.
mittelwert=2.338
bereich=0.002
i=mittelwert-bereich
anzahl=4  
while (i<=mittelwert+bereich):
    
    epsilon=i
    plt.title("Epsilon = %.4f" %epsilon)
    x,y=num(x0, psi_0, psi_h, h, m)
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('$\psi(x)$')
    plt.show()
    i=i+(2*bereich/anzahl)
```

```python
# hier werden um einen Mittelwert von epsilon im Bereich von Mittelwert-bereich bis Mittelwert+bereich Plots 
# in der Quantität anzahl (+-1 durch mögliche Rundungen) mit gleichmäßigen Abständen der epsilon geplottet.
mittelwert=4.088
bereich=0.002
i=mittelwert-bereich
anzahl=4 
while (i<=mittelwert+bereich):
    
    epsilon=i
    plt.title("Epsilon = %.4f" %epsilon)
    x,y=num(x0, psi_0, psi_h, h, m)
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('$\psi(x)$')
    plt.show()
    i=i+(2*bereich/anzahl)
```

```python
# hier werden um einen Mittelwert von epsilon im Bereich von Mittelwert-bereich bis Mittelwert+bereich Plots 
# in der Quantität anzahl (+-1 durch mögliche Rundungen) mit gleichmäßigen Abständen der epsilon geplottet.
mittelwert=5.521
bereich=0.002
i=mittelwert-bereich
anzahl=4 
while (i<=mittelwert+bereich):
    
    epsilon=i
    plt.title("Epsilon = %.4f" %epsilon)
    x,y=num(x0, psi_0, psi_h, h, m)
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('$\psi(x)$')
    plt.show()
    i=i+(2*bereich/anzahl)
```

Damit ergeben sich auf 2 Stellen gerundet für die ersten 3 Energie Eigenwerte $\epsilon_n$ auf 2 Stellen gerundet:
$$\epsilon_1=2.34$$
$$\epsilon_2=4.09$$
$$\epsilon_3=5.52$$

```python
# im folgenden Plotten wir die Eigenwerte noch einmal
x=np.array([1,2,3])
eps=np.array([2.34,4.09,5.52])
plt.plot(x,eps)
plt.title('Eigenwerte $\epsilon$ von $\psi(x)$ gegen die Zahl n')
plt.xlabel('n')
plt.ylabel('$\epsilon_n$')
plt.show()
```
