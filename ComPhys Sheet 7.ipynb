## 2.1

Für die  Dimension von $\frac{dN}{dt}=rN(1-N/K)-\frac{BM^2}{A^2+N^2}$ (2.3) gilt:
$$\left[\frac{dN}{dt}\right]=\frac{Anzahl}{Zeit}$$
Im folgenden stehen die dafür notwendigen Dimensionen der Konstanten aus (2.3) (woebi N, r, K nach VL klar sind)


<math>\begin{align}
    [N]&=Anzahl\\
    [r]&=\frac{1}{Zeit}\\
    [K]&=Anzahl\\
    [A]&=Anzahl\\
    [B]&=\frac{Anzahl}{Zeit}\\
    \end{align}
</math>

Im folgenden Substituieren wir jetzt:
<math>\begin{align}
    n &=\frac{N}{K}\\
    \tau &=\frac{B t}{A}\\
    g&=\frac{A}{K}\\
    h&=\frac{rA}{B}\\
    \end{align}
</math>
Offensichtlich sind alle größen Dimensionslos. Damit erhalten wir für (2.3):
$$\frac{dn}{d\tau}=hn(1-gn)-\frac{n^2}{1+n^2}$$

Jetzt wollen wir die stationären punkte n* berechnen. Dazu muss $\frac{dn}{d\tau}=0$ sein. Wir können nun die Gleicung mit $1+n^2$ multiplizieren un verändern die Nullstellen nicht. Wir erhlaten damit: 
<math>\begin{align}
    0 &=n((n^2+1)h(1-gn)-n)\\
    &=n(-hgn^3+hn^2-hgn+h-n)\\
    &=n(-hgn^3+hn^2-(hg+1)n+h)\\
    \end{align}
</math>
Eine Lösung ist offensichtlich $n^*=0$. Sei also nun $n \neq 0$. Wir können nun durch n dividieren und erhalten eine Kubische GLeichung
<math>\begin{align}
    0 &=-hgn^3+hn^2-(hg+1)n+h\\
    \end{align}
</math>

## 2.2 
Im folgendnen ist $g=\frac{A}{K}$ $g=\frac{1}{7.3}$ nach Aufgabenzettel.

```python
g=1/7.3
```

```python
# Gesamte Funktion der Ableitung mit der Nullstelle 0
def get_func2(n,h):
    return h*n*(1-g*n)-n**2/(1+n**2)
```

```python
# modifizeirte Form der Ableitung mit den selben Nullstellen. (bis auf 0). Polynom dritten Grades
def get_func(n,h):
    return -h*g*n**3+h*n**2-(h*g+1)*n+h
```

In den folgenden 2 Plots ist die Funktion $dn/d\tau$ als Funktion von n zu verschiedenen $h$ geplottet. Im zweiten haben wir die Y-Skala vergrößert. Dort kann man auch die verschiedenen Nullstellen in Abhänigkeit von h ablesen. Wir dachten es es nicht nötig explizit viele Werte herauszuschreiben. In der nächsten Zelle haben wir die Nullstellen für den Wert h=0.55 berechnet.

```python
x0=np.array([0,1,2,4])
print('Nullstellen für h=0.55:', str(sc.fsolve(get_func2, x0, args=(0.55))))
```

```python
# original Plot

n=np.linspace(-0.5,6,10000)
start=0.42
scale=0.01
plt.figure(figsize=(15,10))
for i in range (10):
    plt.plot(n, get_func2(n, (i*scale+start)), label='h= %.2f'%(i*scale+start))
for i in range (10):
    plt.plot(n, get_func2(n, ((i+10)*scale+start)), label='h= %.2f'%((i+10)*scale+start),linestyle='dotted')
plt.plot(n,0*n,linestyle=(0, (1, 5)), label='Null', color='black')
plt.legend(ncol=3)
plt.ylabel(r'$dn/d\tau$')
plt.xlabel('n')
plt.title(r'Diagramm 1: $dn/d\tau$ bei verschiedenen h')
plt.show()
```

Sei 
<math>\begin{align}
     f(n)&=\frac {dn}{d\tau}\\
    \end{align}
</math>
Dann wissen wir nach Vorlesung, dass für einen Fixpunkt $n^*$ mit $f(n^*)=0$ folgende Stabilitätsbedingungen gilt:

i)  Ist $f'(n^*)<0$ so ist $n^*$ ein stabiler Fixpunkt

ii) Ist $f'(n^*)>0$ so ist $n^*$ ein instabiler Fixpunkt

ii) Ist $f'(n^*)=0$ so lässt sich keine Aussage über die Stabilität treffen

Die Ableitung der Funktion bei uns ist gegeben durch

$$-\dfrac{2n}{n^2+1}+\dfrac{2n^3}{\left(n^2+1\right)^2}+h\left(1-gn\right)-ghn$$

```python
def get_func3(n,h):
    return -2*n/(n**2+1)+2*n**3/((n**2+1)**2)+h*(1-g*n)-g*h*n
```

```python
# Plot der 2. Ableitung

n=np.linspace(-0.5,6,10000)
start=0.42
scale=0.01
plt.figure(figsize=(15,10))
for i in range (10):
    plt.plot(n, get_func3(n, (i*scale+start)), label='h= %.2f'%(i*scale+start))
for i in range (10):
    plt.plot(n, get_func3(n, ((i+10)*scale+start)), label='h= %.2f'%((i+10)*scale+start),linestyle='dotted')
plt.plot(n,0*n,linestyle=(0, (1, 5)), label='Null', color='black')
plt.legend(ncol=3)
plt.title(r'Diagramm 2: $d^2n/d\tau^2$ bei verschiedenen h')
plt.ylabel(r'$d^2n/d\tau^2$')
plt.xlabel('n')
plt.show()
```

Wie man an dem letzen Diagramm der 2. Ableitung (bei Vergleich der Nullstellen) sieht, sowie auch schon an den Steigungen in Diagramm 1 und 2 ergeben sich von links nach rechts folgede Aussagen über die Stabilität (sofern die Nullstelle für gegebenes h vorhanden ist):

i) instabil

ii) stabil

iii) instabil

iv) stabil
